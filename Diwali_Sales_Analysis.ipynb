# EXPLORATORY DATA ANALYSIS

 GENDER

```python
# PLOT A BAR CHART FOR GENDER  AND ITS  COUNT
colors = ['Blue', 'Yellow']
ax = sns.countplot(x='Gender',hue = 'Gender', data=df, palette = colors)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# PLOTTING A BAR CHART FOR GENDER VS TOTAL AMOUNT
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sales_gen
```

```python
color = ['Blue','Green']
ax = sns.barplot(x='Gender', y = 'Amount', hue = 'Gender', data=sales_gen, palette = color, legend = False)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```

##
- From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men


## Age

```python
color = ['Blue', 'red']
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', palette = color)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Sales Group
color = ['Blue','Green','Yellow','Orange','Purple','Red','Violet']
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, palette = color, hue = 'Age Group')
```

##
- From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

## State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x='State', y = 'Orders', hue = 'State', data = sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total sales/amount from top 10 states
sales_amou = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(x='State', y='Amount', hue = 'State', data = sales_amou)

for bars in ax.containers:
    ax.bar_label(bars, fmt = '%0.f')
```

##
- From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

## Marital Status

```python
color = ['Blue','Green']
ax = sns.countplot(x='Marital_Status', data = df, palette = color, hue = 'Marital_Status')

sns.set(rc = {'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

##
- From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
sns.set(rc = {'figure.figsize':(20,15)})
ax = sns.countplot(data = df, x='Occupation', palette = 'Set2', hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
amount_occu = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x='Occupation', y = 'Amount', hue = 'Occupation', palette = 'Set1', data = amount_occu)
for bars in ax.containers:
    ax.bar_label(bars, fmt = '%0.f')
```

## Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category', palette = 'Set1')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
ax = sns.barplot(x='Product_Category', y = 'Amount', hue='Product_Category', palette = 'Set2', data = sales_state)
for bars in ax.containers:
    ax.bar_label(bars, fmt = '%0.f')
```

  ##
  - From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category



```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', palette = 'Set2', hue = 'Product_ID')
for bars in ax.containers:
    ax.bar_label(bars)
```
